```python
#converting reservation_status_date from object to datetime
data['reservation_status_date']=pd.to_datetime(data['reservation_status_date'])
```

```python
#dropping unwanted columns

data.drop(['agent','company'],inplace=True,axis=1)
```

```python
#remove outlires from adr coloumn
data=data[data['adr']<5000]
```

##**Analyse the clean data**

```python
cancel_percentage=(data['is_canceled'].value_counts(normalize=True))*100
print(cancel_percentage)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'],data['is_canceled'].value_counts(),width=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=data,palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#resort hoter cancel and not cancelled percentage
resort_hotel=data[data['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#city hoter cancel and not cancelled percentage
city_hotel=data[data['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#any affect on price or not!
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(10,8))
plt.title('Avarage daily rate in city hotel and resort hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='resort hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='city hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
data['month'] = data['reservation_status_date'].dt.month
plt.figure(figsize=(10,8))
ax1=sns.countplot(x='month',hue='is_canceled',data=data,palette='bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title('ADR per month', fontsize=30)
plt.bar('month','adr',data=data[data['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel('month')
plt.ylabel('ADR')
plt.show()
```

```python
canceled_data=data[data['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
data['market_segment'].value_counts(normalize=True)
```

```python
canceled_data['market_segment'].value_counts(normalize=True)
```

```python
not_canceled_data=data[data['is_canceled']==0]
```

```python
canceled_data_adr= canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date', inplace=True)

not_canceled_data_adr= not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(15,8))
plt.title('Avarage daily rate')
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label='not canceled')
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label='canceled')
plt.legend()
plt.show()
```

```python
canceled_data_adr=canceled_data_adr[(canceled_data_adr['reservation_status_date']>'2015-09') & (canceled_data_adr['reservation_status_date']<'2017-09')]
not_canceled_data_adr=not_canceled_data_adr[(not_canceled_data_adr['reservation_status_date']>'2015-09') & (not_canceled_data_adr['reservation_status_date']<'2017-09')]
```
